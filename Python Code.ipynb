```python
#Data Trimming

dftrim = pd.read_csv("Python Session 5/Database_individual_assignment.csv", usecols=[
    "workexperience", "age", "nationality", "gender",
    
    "perfa_t0", "perfb_t0", "perfc_t0", "perfd_t0",
    "perfa_t1", "perfb_t1", "perfc_t1", "perfd_t1",
    "perfa_t2", "perfb_t2", "perfc_t2", "perfd_t2",
    "perfa_t3", "perfb_t3", "perfc_t3", "perfd_t3",
    "perfa_t4", "perfb_t4", "perfc_t4", "perfd_t4",
    
    "enga_t1", "engb_t1", "engc_t1",
    "enga_t2", "engb_t2", "engc_t2",
    "enga_t3", "engb_t3", "engc_t3",
    "enga_t4", "engb_t4", "engc_t4",
    
    "supporta_t1", "supportb_t1", "supportc_t1", "supportd_t1",
    "supporta_t2", "supportb_t2", "supportc_t2", "supportd_t2",
    "supporta_t3", "supportb_t3", "supportc_t3", "supportd_t3",
    "supporta_t4", "supportb_t4", "supportc_t4", "supportd_t4",
    
    "selfa_t0", "selfb_t0", "selfc_t0", "selfd_t0", "selfe_t0",
    "selff_t0", "selfg_t0", "selfh_t0", "selfi_t0", "selfj_t0",
    "selfa_t1", "selfd_t1", "selfe_t1", "selff_t1", "selfg_t1", "selfj_t1",
    "selfa_t2", "selfd_t2", "selfe_t2", "selff_t2", "selfg_t2", "selfj_t2",
    "selfa_t3", "selfd_t3", "selfe_t3", "selff_t3", "selfg_t3", "selfj_t3",
    "selfa_t4", "selfd_t4", "selfe_t4", "selff_t4", "selfg_t4", "selfj_t4"
])
```

```python
dftrim
```

```python
# Data cleaning using listwise deletion

print(dftrim.dtypes)
```

```python
# Checking the count of missing values

print(dftrim.isnull().sum(axis=0))
```

```python
dftrim.replace(' ', 'Missing', inplace=True)
dftrim.replace('Missing',np.nan,inplace=True)
dftrim
```

```python
dftrim.isnull().sum(axis = 0)
```

```python
dftrim= dftrim.dropna()
dftrim
```

```python
# Convert the relevant columns to numeric, handling errors
for col in dftrim.columns:
    try:
        dftrim[col] = pd.to_numeric(dftrim[col])
    except ValueError:
        print(f"Warning: Column '{col}' could not be converted to numeric and will be skipped.")
      
# Calculate average performance
dftrim['avg_perfor'] = dftrim[[
    "perfa_t0", "perfb_t0", "perfc_t0", "perfd_t0",
    "perfa_t1", "perfb_t1", "perfc_t1", "perfd_t1",
    "perfa_t2", "perfb_t2", "perfc_t2", "perfd_t2",
    "perfa_t3", "perfb_t3", "perfc_t3", "perfd_t3",
    "perfa_t4", "perfb_t4", "perfc_t4", "perfd_t4"
]].mean(axis=1)

# Calculate average engagement
dftrim['avg_engagement'] = dftrim[[
    "enga_t1", "engb_t1", "engc_t1",
    "enga_t2", "engb_t2", "engc_t2",
    "enga_t3", "engb_t3", "engc_t3",
    "enga_t4", "engb_t4", "engc_t4"
]].mean(axis=1)

# Calculate average social support
dftrim['avg_socsup'] = dftrim[[
    "supporta_t1", "supportb_t1", "supportc_t1", "supportd_t1",
    "supporta_t2", "supportb_t2", "supportc_t2", "supportd_t2",
    "supporta_t3", "supportb_t3", "supportc_t3", "supportd_t3",
    "supporta_t4", "supportb_t4", "supportc_t4", "supportd_t4"
]].mean(axis=1)


# Calculate average self-efficacy
dftrim['avg_selfefficacy'] = dftrim[[
    "selfa_t0", "selfb_t0", "selfc_t0", "selfd_t0", "selfe_t0",
    "selff_t0", "selfg_t0", "selfh_t0", "selfi_t0", "selfj_t0",
    "selfa_t1", "selfd_t1", "selfe_t1", "selff_t1", "selfg_t1", "selfj_t1",
    "selfa_t2", "selfd_t2", "selfe_t2", "selff_t2", "selfg_t2", "selfj_t2",
    "selfa_t3", "selfd_t3", "selfe_t3", "selff_t3", "selfg_t3", "selfj_t3",
    "selfa_t4", "selfd_t4", "selfe_t4", "selff_t4", "selfg_t4", "selfj_t4"
]].mean(axis=1)

# Display the updated dataframe
dftrim
```

```python
p1 = Process( data=dftrim,model=7,x="avg_selfefficacy",y="avg_perfor",w="avg_socsup",m=["avg_engagement"],controls=["workexperience", "age", "nationality", "gender"],center=True,boot=10000)
p1.summary()
```

```python
g1 = p1.plot_conditional_indirect_effects(med_name="avg_engagement",x="avg_socsup")
g1.add_legend(title="")
plt.show()
```
